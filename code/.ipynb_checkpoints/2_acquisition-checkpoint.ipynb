```python
roi_template_512 = load_hdf5(directories['data'] + 'SMP_M1_RR2_BC2_HD.h5')['ROIs']

roi_template_32 = load_hdf5(directories['data'] + 'SMP_M1_RR2_BC2_HD.h5')['ROIs']
n_roi = roi_template_32.min()
roi_template_32 /= n_roi
roi_template_32 = resize(roi_template_32, (32, 32), 
                     preserve_range = True, order = 0, mode = 'reflect')
roi_template_32 *= n_roi
roi_template_32 = roi_template_32.astype(int)
```

```python
i_rois = np.unique(np.hstack([roi_template_512.flatten(), roi_template_32.flatten()]))
i_rois = i_rois[i_rois != 1]
n_rois = len(i_rois)
roi_colours = sns.color_palette('hls', n_rois)
```

```python
trajectory = np.loadtxt(directories['home'] + '/misc/pwStimBufData_1os.txt')

blanking_mask = trajectory[2, :] != 0

x, y = roi_template_512.shape

x_pixel = np.abs(trajectory[0, blanking_mask][:, None] - np.linspace(-0.5, 0.5, x)[None, :]).argmin(axis = 1)
y_pixel = np.abs(trajectory[1, blanking_mask][:, None] - np.linspace(-0.5, 0.5, y)[None, :]).argmin(axis = 1)

roi_pixel = np.empty(trajectory.shape[1])

for itx in range(x):
    for ity in range(y):
        roi_pixel[(x_pixel == itx) & (y_pixel == ity)] = roi_template_512[itx, ity] 
```

```python
first_spiral = trajectory[:2, :1600]
second_spiral = trajectory[:2, 2000:3600]
```

```python
sns.set_style('white')

fig, ax = plt.subplots(1, 1)

ax.scatter(first_spiral[0, :], first_spiral[1, :], s = 0.1, c = 'k')
ax.scatter(second_spiral[0, :], second_spiral[1, :], s = 0.1, c = 'r', alpha = 0.5)

ax.axis('equal')
ax.set_xlim(-0.45, 0.45)
ax.set_ylim(-0.45, 0.45)
ax.set_xticklabels([]);
ax.set_yticklabels([]);

sns.despine(left = True, bottom = True, ax = ax)

fig.set_size_inches(2, 2)

plt.savefig(directories['figs'] + 'acquisition/' + 'spiral_trajectory.svg', **fig_p)
```

```python
sns.set_style('white')

fig, ax = plt.subplots(1, 1)

ax.scatter(trajectory[0, blanking_mask], trajectory[1, blanking_mask], s = 0.5, c = 'k', alpha = 0.2)

for itx, i_roi in enumerate(i_rois):
    roi_bool = roi_pixel == i_roi
    ax.scatter(
        trajectory[0, blanking_mask][roi_bool], 
        trajectory[1, blanking_mask][roi_bool], 
        color = roi_colours[itx],
        s = 1
    )
    
ax.axis('equal')
ax.set_xlim(-0.45, 0.45)
ax.set_ylim(-0.45, 0.45)
ax.set_xticklabels([]);
ax.set_yticklabels([]);

sns.despine(left = True, bottom = True, ax = ax)

fig.set_size_inches(2, 2)

plt.savefig(directories['figs'] + 'acquisition/' + 'spiral_roi.png', **fig_p)
```
